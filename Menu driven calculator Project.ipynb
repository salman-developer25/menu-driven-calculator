# Menu Driven Calculator

## Project Description

This is a Menu Driven Calculator built using Python.
It allows users to perform basic arithmetic operations through a simple menu interface.

The program runs continuously until the user chooses to exit.

## How It Works

The program displays a menu.

The user selects an operation by entering a specific letter.

The corresponding function is called.

The result is displayed.

The program continues until the user enters E to exit

```python
def addition():
    first_number = int(input("Enter your first_number: "))
    second_number = int(input("Enter your second_number: "))
    Addition = first_number+second_number
    print(f"The addition of {first_number} and {second_number} is {Addition}")
    
def subtraction():
    first_number = int(input("Enter your first_number: "))
    second_number = int(input("Enter your second_number: "))
    Subtraction = first_number-second_number
    print(f"The subtraction of {first_number} and {second_number} is {Subtraction}")

def multiplication():
    first_number = int(input("Enter your first_number: "))
    second_number = int(input("Enter your second_number: "))
    Multiplication = first_number*second_number
    print(f"The multiplication of {first_number} and {second_number} is {Multiplication}")

def division():
    first_number = int(input("Enter your first_number: "))
    second_number = int(input("Enter your second_number: "))
    
    if second_number == 0:
        print("Error: Division by zero is not allowed.")
    else:
        Division = first_number/second_number
        print(f"The division of {first_number} and {second_number} is {Division}")

def floor_division():
    first_number = int(input("Enter your first_number: "))
    second_number = int(input("Enter your second_number: "))
    Floor = first_number//second_number
    print(f"The floor division of {first_number} and {second_number} is {Floor}")

def modulus():
    first_number = int(input("Enter your first_number: "))
    second_number = int(input("Enter your second_number: "))
    Modulus = first_number%second_number
    print(f"The modulus of {first_number} and {second_number} is {Modulus}")
    
def power():
    number = int(input("Enter the number you want find it exponent: "))
    Power = number**2
    print(f"The power of {number} is {Power}")
    
def double_num():
    number = int(input("Enter the number to double it: "))
    Double = number*2
    print(f"The double of {number} is {Double}")



while (True):
    calculation = input("Menu Driven Calculator\n A for Addition\n S for Subtration\n M for Multipliction\n D for Divsion\n F for Floor division \n MO for Modulus\n P for Power\n DN for Double\n E for Exist").upper()
    if calculation == "A":
        addition()
    elif calculation == "S":
        subtraction()
    elif calculation == "M":
        multiplication()
    elif calculation == "D":
        division()
    elif calculation == "F":
        floor_division()
    elif calculation == "MO":
        modulus()
    elif calculation == "P":
        power()
    elif calculation == "DN":
        double_num()
    elif calculation == "E":
        break
    else:
        print("Wrong input please enter the alphabits that is given")
```
